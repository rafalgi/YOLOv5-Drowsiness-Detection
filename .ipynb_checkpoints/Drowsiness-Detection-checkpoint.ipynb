# Instal and import dependence

```python
import torch
import matplotlib.pyplot as plt
import cv2
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

# Load Model

```python
model = torch.hub.load('ultralytics/yolov5' , 'yolov5s')
```

```python
model
```

# Make Detections

```python
img = cv2.imread('pies.jpg')
```

```python
result = model(img)
result.print()
```

```python
%matplotlib inline 
plt.imshow(np.squeeze(result.render()))
plt.show()
```

```python
# Wczytanie obrazw
images = [
    cv2.imread('kot.jpg'),
    cv2.imread('kot1.jpg'),
    cv2.imread('kot3.jpg'),
    cv2.imread('psy1.jpg')
]

# Utworzenie figury i osi (subplotw) w siatce 2x2
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Iteracja przez osie i obrazy, aby je wywietli
for ax, img in zip(axes.flat, images):
    ax.imshow(img)

# Wywietlenie wykresu
plt.show()
```

```python
# Przetwarzanie obrazw za pomoc modelu YOLOv5
results = [model(img) for img in images]

# Utworzenie figury i osi (subplotw) w siatce 2x2
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Iteracja przez osie i przetworzone obrazy, aby je wywietli
for ax, result in zip(axes.flat, results):
    # Renderowanie obrazu z zaznaczonymi obiektami
    img = np.squeeze(result.render())
    
    ax.imshow(img)  # Wywietlenie obrazu (w formacie BGR)
    ax.axis('off')  # Wyczenie osi dla estetyki

# Wywietlenie caoci
plt.show()
```

# Real time Detections

```python
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret , frame = cap.read()

    result = model(frame)
    cv2.imshow('Camera' , np.squeeze(result.render()))

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

# Train from scratch

```python
import uuid
import os
import time
```

```python
IMAGES_PATH = os.path.join('data' , 'images')
labels  = ['awake' , 'drowsy']
number_imgs = 20
```

```python
cap = cv2.VideoCapture(0)
# Loop through labels
for label in labels:
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    
    # Loop through image range
    for img_num in range(number_imgs):
        print('Collecting images for {}, image number {}'.format(label, img_num))
        
        # Webcam feed
        ret, frame = cap.read()
        
        # Naming out image path
        imgname = os.path.join(IMAGES_PATH, label+'.'+str(uuid.uuid1())+'.jpg') # gdzie maj by zapisywane screeny / nazwaa obrazu z id
        
        # Writes out image to file 
        cv2.imwrite(imgname, frame)
        
        # Render to the screen
        cv2.imshow('Image Collection', frame)
        
        # 2 second delay between captures
        time.sleep(2)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
